```python
print("=== Program CRUD Sederhana ===")
print("====== Tasks Management ======")
print("===== Ahmad Fajril Falah =====")

tasks = []

while True:
    print("\nPILIH MENU")
    print("1. Read  (Lihat Tugas)")
    print("2. Create (Tambah Tugas)")
    print("3. Update (Edit Tugas)")
    print("4. Delete (Hapus Tugas)")
    print("5. Keluar")

    pilih = input("Pilih menu (1-5): ")

    # READ
    if pilih == "1":
        print("\nDAFTAR TUGAS:")
        if len(tasks) == 0:
            print("Belum ada tugas.")
        else:
            for i in range(len(tasks)):
                print(f"{i+1}. {tasks[i]}")

    # CREATE
    elif pilih == "2":
        tugas = input("Masukkan tugas baru: ")
        tasks.append(tugas)
        print("Tugas berhasil ditambahkan!")

    # UPDATE
    elif pilih == "3":
        if len(tasks) == 0:
            print("Belum ada tugas.")
        else:
            for i in range(len(tasks)):
                print(f"{i+1}. {tasks[i]}")
            index = int(input("Pilih nomor tugas yang ingin diubah: ")) - 1
            tugas_baru = input("Masukkan tugas baru: ")
            tasks[index] = tugas_baru
            print("Tugas berhasil diupdate!")

    # DELETE
    elif pilih == "4":
        if len(tasks) == 0:
            print("Belum ada tugas.")
        else:
            for i in range(len(tasks)):
                print(f"{i+1}. {tasks[i]}")
            index = int(input("Pilih nomor tugas yang ingin dihapus: ")) - 1
            print(f"Tugas '{tasks[index]}' dihapus.")
            tasks.pop(index)

    # EXIT
    elif pilih == "5":
        print("Program selesai.")
        break

    else:
        print("Pilihan tidak valid!")
```
